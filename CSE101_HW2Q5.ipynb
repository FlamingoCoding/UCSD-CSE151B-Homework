```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

def confirmTriangle(A):
    n = len(A)
    for v in range(n):
        for w in range(v + 1, n):
            if A[v][w] == 1:
                for u in range(n):
                    if u != v and u != w:
                        if A[u][v] == 1 and A[u][w] == 1:
                            return True
    return False

def generate_graph_with_density(n, p):
    """
    生成一个 n 节点的无向图，边以概率 p 出现。
    """
    A = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                A[i][j] = A[j][i] = 1
    return A

# 实验参数
n = 256                     # 固定节点数
densities = [0.1, 0.3, 0.5, 0.7, 0.9]   # 不同的边密度
trials = 50                 # 每种密度下重复次数

avg_times = []

# 运行实验
for p in densities:
    times = []
    for _ in range(trials):
        A = generate_graph_with_density(n, p)
        start = time.perf_counter()
        confirmTriangle(A)
        times.append(time.perf_counter() - start)
    avg_times.append(np.mean(times))
    print(f"密度 p={p:.1f} 的平均运行时间：{avg_times[-1]:.4f} 秒")

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(densities, avg_times, marker='o', linestyle='-', linewidth=2, label=f'n = {n}')

# 在每个点上标出坐标
for p, t in zip(densities, avg_times):
    plt.text(p, t, f"({p:.1f}, {t:.4f}s)",
             fontsize=9, ha='center', va='bottom')

plt.xlabel('边密度 p')
plt.ylabel('平均运行时间 (秒)')
plt.title(f'节点数 = {n} 时，不同边密度下三角形检测运行时间')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
